```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import matplotlib
from plotly.subplots import make_subplots
import seaborn as sns

```

```python
def get_losses(name):
    losses = []
    for i in [2,5,8,11,14,17,20]:
        loss = np.loadtxt("../saved_results/synthetic_biological_trussart_uniform_trans_conf/trans/synthetic_biological_trussart_uniform_trans_conf_trans_" + name + str(i) + ".txt")
        losses.append(loss)
    return losses
```

```python
general_loss = get_losses('losses_')
```

```python
train_bio_loss = get_losses('train_biological_losses_all_epochs_')
test_bio_loss = get_losses('test_biological_losses_all_epochs_')
```

```python
train_kabsch_loss = get_losses('train_kabsch_losses_all_epochs_')
test_kabsch_loss = get_losses('test_kabsch_losses_all_epochs_')
```

```python
train_distance_loss = get_losses('train_distance_losses_all_epochs_')
test_distance_loss = get_losses('test_distance_losses_all_epochs_')
```

```python
trussart_loss = get_losses('trussart_test_kabsch_losses_all_epochs_')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 10))
labels = [r'$\lambda_k = 0$ $\lambda_b = 0$', r'$\lambda_k = 0.1$ $\lambda_b = 0$', r'$\lambda_k = 0$ $\lambda_b = 0.1$',  r'$\lambda_k = 0.1$ $\lambda_b = 0.1$',  r'$\lambda_k = 1$ $\lambda_b = 0.1$',  r'$\lambda_k = 0.1$ $\lambda_b = 1$',  r'$\lambda_k = 1$ $\lambda_b = 1$']

for i in range(len(general_loss)):
    axs[0,0].plot(general_loss[i], color = sns.color_palette("Set2")[i])
    axs[0,1].plot(train_bio_loss[i], label=labels[i], linestyle = "-", color = sns.color_palette("Set2")[i])
    axs[0,1].plot(test_bio_loss[i], linestyle = "--", color = sns.color_palette("Set2")[i])
    axs[1,0].plot(train_kabsch_loss[i], label=labels[i], linestyle = "-", color = sns.color_palette("Set2")[i])
    axs[1,0].plot(test_kabsch_loss[i], linestyle = "--", color = sns.color_palette("Set2")[i])
    axs[1,1].plot(trussart_loss[i], label=labels[i], color = sns.color_palette("Set2")[i])


axs[0,0].legend(labels, loc = (0.2, -1.5), ncol=4)
axs[0,1].legend(["Train","Test"], loc = (0.3, -1.46), ncol=2)
axs[0,0].title.set_text("(a) General Loss")

axs[0,0].set_ylabel("Loss")
axs[0,0].set_xlabel("Epoch")

axs[0,1].set_ylabel("Loss")
axs[0,1].set_xlabel("Epoch")

axs[1,0].set_ylabel("Kabsch Distance")
axs[1,0].set_xlabel("Epoch")

axs[1,1].set_ylabel("Kabsch Distance")
axs[1,1].set_xlabel("Epoch")

axs[0,1].title.set_text("(b) Biological Loss")
axs[1,0].title.set_text("(c) Kabsch Distance Synthetic Loss")
axs[1,1].title.set_text("(d) Kabsch Distance Trussart Loss")

leg = axs[0,1].get_legend()
leg.legendHandles[0].set_color('black')
leg.legendHandles[1].set_color('black')

plt.show()

```
